```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('ROPCond_Stage2.jpg',0)
plt.figure(figsize=(10,10))
plt.imshow(img, cmap='gray')
```

```python
output=cv2.equalizeHist(img)
plt.imshow(output,cmap='gray');
plt.show()
```

```python
thresh1 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
										cv2.THRESH_BINARY, 47, 6)

thresh2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
										cv2.THRESH_BINARY, 19, 4)

thresh3 = cv2.adaptiveThreshold(output, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
										cv2.THRESH_BINARY, 11, 1)
# 47 7
# 67 5
# 97 7
# 19 4
cv2.imshow('Adaptive Gaussian47-6', thresh1)
cv2.imshow('Adaptive Gaussian67-5', thresh2)
cv2.imshow('Adaptive Gaussian19-4', thresh3)
cv2.waitKey(0)
```

```python
thresh = 120
im_bw = cv2.threshold(thresh1, thresh, 255, cv2.THRESH_BINARY)[1]
plt.imshow(im_bw,cmap='gray')
```

```python
plt.imshow(thresh3,cmap='gray')
```

```python
kernel = np.ones((2, 2), np.uint8)
kernel2 = np.ones((2, 1), np.uint8)

img_dilation = cv2.dilate(thresh3, kernel, iterations=2)
img_erosion = cv2.erode(img_dilation, kernel2, iterations=2)

cv2.imshow('Input', thresh3)
cv2.imshow('Erosion', img_erosion)
cv2.imshow('Dilation', img_dilation)

cv2.waitKey(0)
```
