Take a random image, then draw its histogram and apply image enhancement techniques to it.  (Practical)

```python
import matplotlib.pyplot as plt
import cv2 
import numpy as np
```

```python
im = plt.imread("Clownfish.jpeg")
plt.imshow(im);
```

```python
# Using 0 to read image in grayscale mode
im = cv2.imread('Clownfish.jpeg',0)

plt.hist(im.ravel(),255,[0,255])
plt.show()
cv2.imshow("img",im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#inversing image
im_g = cv2.imread('Clownfish.jpeg',0)
im_inverse = 255 - im_g
plt.imshow(im_inverse,cmap="gray");
```

```python
plt.hist(im_inverse.ravel(),255,[0,255])
plt.show()
```

```python
output=cv2.equalizeHist(im)
plt.imshow(output,cmap='gray');
plt.show()
plt.hist(output.ravel(),255,[0,255])
plt.show()
```

```python
im = cv2.imread('Clownfish.jpeg',0)
im_log_law_transformation = 0.6 * (np.log(1 + np.float32(im)))
plt.imshow(im_log_law_transformation,cmap='gray');
plt.show()
plt.hist(im_log_law_transformation.ravel(),255,[0,255])
plt.show()
```
