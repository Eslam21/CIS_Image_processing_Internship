```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('Retinopathye.jpg',0)
plt.figure(figsize=(10,10))
plt.imshow(img, cmap='gray')
```

```python
img = cv2.imread('Retinopathye.jpg')
b,g,r = cv2.split(img)

f, axarr = plt.subplots(1,3,figsize=(15,15))

axarr[0].imshow(r)
axarr[1].imshow(g)
axarr[2].imshow(b)
```

```python
thresh1 = cv2.adaptiveThreshold(g, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
										cv2.THRESH_BINARY, 221, 6)
plt.figure(figsize=(10,10))
plt.imshow(thresh1,cmap='gray')
```
